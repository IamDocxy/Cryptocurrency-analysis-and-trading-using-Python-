```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('coinbaseUSD_1-min_data_2014-12-01_to_2018-01-08.csv')
```

```python
df.tail()
```

# Converting Timestamp to Date Time Object
### Adding another column to store datetime objects. Might need timestamps later

```python

df['Date'] = pd.to_datetime(df['Timestamp'],unit='s') 
```

```python
df.columns
```

# Can't find any null values. So dataset is complete

```python
 df.isnull().describe()
```

# Now lets check the trading volume's relation with the Closing price.
#### I am extremely curious on how it effects the closing prices. Lets find out

```python
plt.figure(figsize=(18,10))
plt.scatter(x='Volume_(Currency)',y='Close',data=df,marker='o')
plt.ylabel('Closing price')
plt.xlabel('24 Hour volume')
plt.legend()
```
